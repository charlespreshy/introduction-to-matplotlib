### Introduction to matplotlib

Matplotlib is one of the most popular Python packages used for data visualisation. It is a cross-platform library for making 2D plots from data in arrays. It provides an object-oriented API that helps in embeddig plots in applications using python.

It works together with Numpy to build Graphical Representation for data.

The major modules of the matplotlib library used for plotting graphs is the pyplot.

Matplotlib.pyplot is a collection of command style functions that make Matplotlib work like MATLAB. Each pyplot function makes some change to a figure. For example, a function creates a figure, a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

### Important function used in pyplot

###### Image functions

- imread: used to read image from a file into an array
- imsave: used to save an array as an image file
- imshow: used to display an image on the axes

###### Axis function

- Axes(): used to add axes to the figure
- text(): adds text to the axes
- title(): adds title to the selected axes
- xlabel(): sets the x axia label
- ylabel(): sets the y axis label

###### Figure function
- figtext(): adds text to the figure
- figure(): creates a  new figure
- savefig(): save the current figure
- show(): display a figure
- close(): close the figure window

Figure:It is a whole figure which may contain one or more than one axes(plots). You can think of a figure as a canvas which contains plots.

Axes: It is what we generally think of as a plot. A figure can contaiun many Axes. It contains two or three (in the case of 3D)Axix objects. Each Axes has a title, an x-label and a y-label.

Axis: They are the number line like objects and take care of generating the graph limits.

```python
x = [2,4,5]
y = [4,8,10]

#set figure size
plt.figure(figsize = (10,5))

plt.plot(x,y)
plt.title("My simple plot")
plt.xlabel("Weights (kg)")
plt.ylabel("Accelertion (m/s2kg)")

plt.show()
```

```python
from matplotlib import style

style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'g',label='milo', linewidth = 5)
plt.plot(x2,y2,'c',label='Bournvita', linewidth = 5)

plt.title('Commodity Sales over time')
plt.ylabel('Price (N1000)')
plt.xlabel('Product (cartons)')

plt.legend()

plt.grid(True,color = 'k')

plt.show()
```
